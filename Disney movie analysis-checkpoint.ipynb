## Task 4: Describe the total counts of Genres

```python
plt.subplots(figsize=(10,10))
data.groupby(['genre'])['total_gross'].sum().sort_values(ascending = False).head(20).plot(kind = 'pie')
plt.title('Disney Movie Analysis', fontsize = 24)
plt.xlabel('genre', fontsize = 18)
plt.ylabel('total_gross',fontsize = 18)
plt.xticks(rotation = 90)
plt.show()
```
